# Experimentation with Seaborn Visualization #

Using the Seaborn Tutoral from: https://www.youtube.com/watch?v=6GUZXDef2U0

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())

crash_df = sns.load_dataset('car_crashes')
crash_df
```

## Distribution Plot ##

```python

sns.displot(crash_df['not_distracted'], kde='True', bins = 25)
#sns.histplot(crash_df['not_distracted'])
```

## Joint Plot ##

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind = 'scatter')
plt.show()

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind = 'hex')
```

## KDE Plot

```python
sns.kdeplot(crash_df['alcohol'])
```

## Pair Plots ##

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.pairplot(tips_df, hue='sex', palette='Blues')
```

## Rug Plot ##

```python
sns.rugplot(tips_df['tip'])
```

## Styling ##

```python
sns.set_style('white')
#sns.set_style('darkgrid')
#sns.set_style('whitegrid')
#sns.set_style('dark')
#sns.set_style('ticks')
plt.figure(figsize=(8,4))

sns.set_context('poster', font_scale=1.4)
sns.jointplot(x = 'speeding', y = 'alcohol', data=crash_df, kind ='reg')
plt.show()

sns.set_context('talk', font_scale=1.4)
sns.jointplot(x = 'speeding', y = 'alcohol', data=crash_df, kind ='reg')
plt.show()

sns.set_context('paper', font_scale=1.4)
sns.jointplot(x = 'speeding', y = 'alcohol', data=crash_df, kind ='reg')
```

# Categorical Plots #

## Bar Plots 

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.cov)
```

## Count Plot

```python
sns.countplot(x = 'sex', data=tips_df)
```

## Box Plot

```python
sns.boxplot(x='day', y='total_bill', data = tips_df, hue='sex')

plt.legend(loc=0)
```

## Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex')
plt.show()

sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split =True)
```

## Strip Plot

```python
plt.figure(figsize=(8,5))

sns.stripplot(x='day', y='total_bill', data = tips_df, jitter=True)
plt.show()

sns.stripplot(x='day', y='total_bill', data = tips_df, jitter=True, hue='sex')
plt.show()

sns.stripplot(x='day', y='total_bill', data = tips_df, jitter=True, hue='sex', dodge=True)
```

## Swarm Plot

```python
sns.violinplot(x='day', y='total_bill', data = tips_df)

sns.swarmplot(x='day', y='total_bill', data = tips_df, color ='white')
plt.show()

sns.swarmplot(x='day', y='total_bill', data = tips_df, hue='sex')
```

## Palettes

```python
sns.set_style('dark')

# Google Matplotlib Colormaps
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette ='magma')
plt.legend(loc=2)
plt.show()

sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette ='afmhot')

#"best=0, upper right = 1, upper left = 2, lower left = 3, lower right = 4
# matplotlib pyplot legend examples
plt.legend(loc=0)
plt.show()
```

# Matrix Plots #
## Heatmaps ##

```python
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale =1.2)

# numeric_only is new from tutorial
crash_mx = crash_df.corr(numeric_only=True)

# make sure seaborn is pip install --upgraded
sns.heatmap(crash_mx, annot=True, cmap='Blues')
plt.show()
```

```python
flights = sns.load_dataset('flights')

flights_pt = flights.pivot_table(index = 'month', columns='year', values='passengers')
flights_pt

sns.heatmap(flights_pt, cmap='Blues', linecolor='white', linewidth=1)
```

## Cluster Map

```python
sns.clustermap(flights_pt, cmap='Blues', standard_scale=1)
```

## PairGrid

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue='species')


iris_g.map_diag(plt.hist)
#iris_g.map_offdiag(plt.scatter)
iris_g.map_upper(plt.scatter)
iris_g.map_lower(sns.kdeplot)
plt.show()
```

## Facet Grid

```python
tips_fg = sns.FacetGrid(tips_df, col='time', row='smoker')
tips_fg.map(plt.scatter, 'total_bill', 'tip')
plt.show()
```

```python
tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height = 4, aspect=1.3, col_order=['Dinner', 'Lunch'], palette='Set1')
tips_fg.map(plt.scatter, 'total_bill', 'tip', edgecolor='w')
plt.show()
```

```python
kws=dict(s=50, linewidth=.5, edgecolor='w')

tips_fg = sns.FacetGrid(tips_df, col='sex', hue='smoker', height = 4, aspect=1.3, hue_order=['Yes', 'No'], hue_kws=dict(marker=['^', 'v']))
tips_fg.map(plt.scatter, 'total_bill', 'tip', **kws)
plt.show()
```

```python
att_df = sns.load_dataset('attention')

att_fg = sns.FacetGrid(att_df, col='subject', col_wrap=5, height =1.5)

att_fg.map(plt.plot, 'solutions','score', marker='.')
```

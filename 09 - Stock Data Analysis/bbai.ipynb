```python
import yfinance as yf 
import pandas as pd 
import matplotlib.pyplot as plt 
from datetime import datetime
```

```python
df_close = yf.download("BBAI")['Close']
df_close = df_close.tz_localize(None)
# df['Date'] = pd.to_datetime(df['Date'])
# df = df.set_index('Date')
print(df_close.index)
print(df_close)
df_close.head()
```

```python
df_open = yf.download("BBAI")['Open']
df_open = df_open.tz_localize(None)
# df['Date'] = pd.to_datetime(df['Date'])
# df = df.set_index('Date')
df_open.head()
```

```python
df_high = yf.download("BBAI")['High']
df_high = df_high.tz_localize(None)
# df['Date'] = pd.to_datetime(df['Date'])
# df = df.set_index('Date')
df_high.head()
```

```python
df_low = yf.download("BBAI")['Low']
df_low = df_low.tz_localize(None)
# df['Date'] = pd.to_datetime(df['Date'])
# df = df.set_index('Date')
df_low.head()
```

```python
# Define the date range
start_date = pd.Timestamp("2024-10-15")  # Add the UTC timezone
end_date = pd.Timestamp(datetime.now())  # Add the UTC timezone

# Filter the Series
filtered_close = df_close[(df_close.index >= start_date) & (df_close.index <= end_date)]
filtered_close_df = filtered_close.reset_index()
filtered_close_df.columns = ["Date", "Close"]  # Rename columns for clarity

filtered_open = df_open[(df_open.index >= start_date) & (df_open.index <= end_date)]
filtered_open_df = filtered_open.reset_index()
filtered_open_df.columns = ["Date", "Open"]  # Rename columns for clarity

filtered_high = df_high[(df_high.index >= start_date) & (df_high.index <= end_date)]
filtered_high_df = filtered_high.reset_index()
filtered_high_df.columns = ["Date", "High"]  # Rename columns for clarity

filtered_low = df_low[(df_low.index >= start_date) & (df_low.index <= end_date)]
filtered_low_df = filtered_low.reset_index()
filtered_low_df.columns = ["Date", "Low"]  # Rename columns for clarity
print(filtered_close)
print(filtered_open)
```

```python
sales = pd.read_csv("./data/BBAI Holdings Sales.csv")
sales.head()
```

```python
# Merge all together
sales["Date"] = pd.to_datetime(sales["Date of Sale"])
```

```python
# Merge on the "Date" column
sales = pd.merge(sales, filtered_close_df, on="Date", how="inner")
sales = pd.merge(sales, filtered_open_df, on="Date", how="inner")
sales = pd.merge(sales, filtered_high_df, on="Date", how="inner")
sales = pd.merge(sales, filtered_low_df, on="Date", how="inner")
```

```python
sales.head()
```

```python
# Plot the data
plt.figure(figsize=(10, 6))

# Plot each column
plt.plot(sales['Date'], sales['cost'], label='Sale Price')
plt.plot(sales['Date'], sales['Close'], label='Close')
plt.plot(sales['Date'], sales['Open'], label='Open')
plt.plot(sales['Date'], sales['High'], label='High')
plt.plot(sales['Date'], sales['Low'], label='Low')

# Customize the plot
plt.xlabel("Date")
plt.ylabel("Prices")
plt.title("BBAI Sale Price, Open and Close")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

# Show the plot
plt.tight_layout()
plt.show()
```

```python
sales['beat_open'] = sales['cost'] - sales['Open']
sales['beat_close'] = sales['cost'] - sales['Close'] 
sales['beat_high'] = sales['cost'] - sales['High']
sales['beat_low'] = sales['cost'] - sales['Low']
print("BBAI Ultimate holdings beat the Open by: " + str(sales['beat_open'].mean()))
print("BBAI Ultimate holdings beat the Close by: " + str(sales['beat_close'].mean()))
print("BBAI Ultimate holdings beat the High by: " + str(sales['beat_high'].mean()))
print("BBAI Ultimate holdings beat the Low by: " + str(sales['beat_low'].mean()))
```
